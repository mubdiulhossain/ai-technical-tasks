```python
from llama_cpp import Llama
```

```python
#model_name = "TinyLlama/TinyLlama-1.1B-Chat-v1.0"
MODEL_PATH = "models/Phi-3-mini-4k-instruct-q4.gguf"

llm = Llama(model_path=MODEL_PATH)
```

```python
company = {
    "name": "Betopia Limited",
    "industry": "IT services and consulting",
    "size": "500 employees",
    "challenges": "Improving internal productivity and automating support tasks",
    "website": "https://betopia.com"
}

persona = {
    "name": "Jane Smith",
    "role": "Head of Operations, Boogle",
    "bio": "10+ years experience in IT operations, focused on efficiency, team productivity, and process automation.",
    "responsibilities": "Overseeing internal processes, improving team efficiency, and implementing automation tools"
}

product = {
    "name": "Betopia Assist",
    "description": "Virtual AI assistant that helps customers find products, answers queries, and enhances the online shopping experience."
}
```

```python
def build_prompt(company, persona, product, tone="friendly", length="200"):
    return f"""
You are an AI that writes sales outreach emails.

Write a {tone}, no more than {length}-word email from {company['name']} ({company['industry']}) to {persona['name']}.
Use ONLY the information below.

Recipient ({persona['name']}):
- Role: {persona['role']}
- Bio: {persona['bio']}
- Responsibilities: {persona['responsibilities']}

{company['name']} Product:
- {product['name']}: {product['description']}

Follow this:
- Start directly with "Dear {persona['name']},"
- Do NOT invent facts, numbers, or metrics.
- Do NOT use placeholders.
- Do NOT explain how {product['name']} works; only mention it can help their company or users.
- Do NOT include a subject line, attachments, templates, extra sections, or "body:".
- Include a section explaining why {product['name']} is a strong match for {persona['name']}, based ONLY on the information above.
- Include a sign-off by saying "Best Regards"
"""
```

```python
prompt = build_prompt(company, persona, product, tone="friendly", length="200")
```

```python
response = llm(
    prompt,
    max_tokens=500,  # Adjust to fit ~200 words
    temperature=0.4,  # Creativity/friendliness
    stop = ["Best Regards"]
)
```

```python
email_text = response["choices"][0]["text"]
print(email_text)
```
